```python
import astromugs.pipeline as pipeline
import astromugs.dust as dust
from astromugs.plotting import plot
```

```python
drift = pipeline.Interface()
```

```python
thermpath='thermal/'
chempath='chemistry/'
```

```python
drift.add_thermal_path(thermpath)
drift.add_chemical_path(chempath)
```

```python
drift.params.disk.ref_radius = 100 ##
drift.params.disk.tmidplan_ref = 10 ##
drift.params.disk.nr = 101 ##
drift.params.disk.rin = 1
drift.params.disk.rout = 200
drift.params.disk.star_mass = 1.
drift.params.disk.sigma_compute = 'custom'
drift.params.disk.sigma_path = '../basic_100k/dustpy/dustpy_output/surface_densities.txt'
```

```python
drift.params
```

```python
drift.thermalparams.control.itempdecoup = 1
drift.thermalparams.control.nphot_therm = 8e6
drift.thermalparams.control.nphot_scat = 8e6
drift.thermalparams.control.nphot_mono = 8e6
drift.thermalparams.control.scattering_mode_max = 5
drift.thermalparams.control.modified_random_walk = 1

drift.thermalparams.star.mass = drift.params.disk.star_mass
drift.thermalparams.star.luminosity = 1
drift.thermalparams.star.temperature = 4500
```

```python
drift.thermalparams.wave.lmin_mono = 9.12e-2
drift.thermalparams.wave.lmax_mono = 5e-1
drift.thermalparams.wave.nlam_mono = 10
```

```python
drift.grid.set_wavelength_grid(log=True) #microns
drift.grid.set_mcmonowavelength_grid(log=True)
```

```python
# #-----ADD STARS-----
drift.add_star() #note that if you enter this multiple times, it will keep adding stars.
```

```python
# #-----DEFINE SPHERICAL GRID-----
drift.grid.set_spherical_grid()
```

```python
drift.write_continuum(
    control=True,
    stars=True,
    wave=True,
    grid=True,
    mcmono=True
)
```

```python
# the user can use an independent path here because most of the time the dust_size file is somewhere else, for instance in MHD outputs.
d = dust.CustomDistrib(rho_m=1.67, path="../basic_100k/dustpy/dustpy_output/", units='microns') #microns
```

```python
drift.grid.add_dust(d)
```

```python
drift.add_disk(dust=d)
```

```python
drift.write_continuum(
    dens=True
)
```

```python
plot.density2D_grid(path=thermpath, vmin=1e-25, vmax=1e-12, xlim=(0, 190), ylim=(-180, 180), dust=d, cmap='magma', dens_type='number')
```
